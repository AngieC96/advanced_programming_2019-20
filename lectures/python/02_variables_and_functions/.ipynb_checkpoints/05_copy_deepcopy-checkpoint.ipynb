### Shallow is the default

```python
l = [1,2,[4,5,6], 8]
o = l # a sticky note is attacked to l -> they have the same id!
```

```python
o == l
```

```python
o is l
```

```python
id(o), id(l)
```

```python
o = list(l) # I iterate through the list and I create a new list
```

```python
o == l
```

```python
o is l
```

It constructs another element! list() is the CONSTRUCTOR of list, and by definition it returns a NEW object!

```python
l.append(9)
print(l)
print(o)
```

```python
l[2].append(7)
print(l)
print(o)
```

SHALLOW COPY is the default! The inner lists have the same pointer! â†’ Python do the cheapest thing

```python
import copy as cp # copy module
```

```python
o = cp.copy(l)
l[2].append(888)
print(l)
print(o)
print(id(l[2]))
print(id(o[2]))
```

Also with the module copy I got a shallow copy! It's the cheapest thing!\
They are NOT the same object but they contain the same object!

Mutable objects are tricky, mutable objects with mutable objects inside is something to avoid as much as possible!

```python
o = cp.deepcopy(l)
l[2].append(3333)
print(l)
print(o)
```

```python
def augmented(ll):
    ll+=[99] # equal to ll.__iadd__(99) (special method that you invoke with +=)
             # I have two sticky notes to the list [1, 2, [4, 5, 6, 7, 888, 888, 3333], 8, 9]

augmented(l)
print(l)
```

List are passed by reference! l is MODIFIED!!!

```python
def assignment(ll):
    ll = [99] # ll is a sticky note to the list [99]

assignment(l)
print(l)
```
